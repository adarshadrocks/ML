```python
import cv2
import numpy as np

#load the image
normal_image=cv2.imread("cat.jpg")

option='''
Press 1 : To draw a line on image
Press 2 : To draw a rectangle on image
Press 3 : To draw a circle on image
Press 4 : To draw a pentagon on image
Press 5 : To write a text on image
Press 0 : To exit
'''
#print shape
print(normal_image.shape)

choice=1
while choice!='0':
    print(option)
    choice=input("Enter Your Choice : ")
    if choice=='1':
        x1=int(input("Enter the x1 co-ordinate for the line : "))
        y1=int(input("Enter the y1 co-ordinate for the line : "))
        x2=int(input("Enter the x2 co-ordinate for the line : "))
        y2=int(input("Enter the y2 co-ordinate for the line : "))
        blue_color=int(input("Enter the blue color value for line : "))
        green_color=int(input("Enter the green color  value for line : "))
        red_color=int(input("Enter the red color value for line : "))
        line_width=int(input("Enter the width of the line : "))
        #draw the line in image
        image_with_line=cv2.line(normal_image,(x1,y1),(x2,y2),(blue_color,green_color,red_color),line_width)
        #show the image
        cv2.imshow("Image With Line",image_with_line)
        #never close the image until any exit button is pressed
        cv2.waitKey(0)
        #after click any exit button it destroy all windows 
        cv2.destroyAllWindows()
        choice_for_save_image=input("Want to save the image (y/n) : ")
        if choice_for_save_image=='y':
            cv2.imwrite("Image_With_Line.jpg",image_with_line)
        else:
            print("Ok Thanks.....")
    elif choice=='2':
        x1=int(input("Enter the x1 co-ordinate for the rectangle : "))
        y1=int(input("Enter the y1 co-ordinate for the rectangle : "))
        x2=int(input("Enter the x2 co-ordinate for the rectangle : "))
        y2=int(input("Enter the y2 co-ordinate for the rectangle : "))
        blue_color=int(input("Enter the blue color value for line : "))
        green_color=int(input("Enter the green color  value for line : "))
        red_color=int(input("Enter the red color value for line : "))
        line_width=int(input("Enter the width of the line : "))
        #draw the rectangle in image
        image_with_rectangle=cv2.rectangle(normal_image,(x1,y1),(x2,y2),(blue_color,green_color,red_color),line_width)
        #show the image
        cv2.imshow("Image With Rectangle",image_with_rectangle)
        #never close the image until any exit button is pressed
        cv2.waitKey(0)
        #after click any exit button it destroy all windows 
        cv2.destroyAllWindows()
        choice_for_save_image=input("Want to save the image (y/n) : ")
        if choice_for_save_image=='y':
            cv2.imwrite("Image_With_Rectangle.jpg",image_with_rectangle)
        else:
            print("Ok Thanks.....")
    elif choice=='3':
        x=int(input("Enter the x co-ordinate for the circle : "))
        y=int(input("Enter the y co-ordinate for the circle : "))
        circle_radius=int(input("Enter the radius of the circle : "))
        blue_color=int(input("Enter the blue color value for line : "))
        green_color=int(input("Enter the green color  value for line : "))
        red_color=int(input("Enter the red color value for line : "))
        line_width=int(input("Enter the width of the line : "))
        #draw the circle in image
        image_with_circle=cv2.circle(normal_image,(x1,y1),circle_radius,(blue_color,green_color,red_color),line_width)
        #show the image
        cv2.imshow("Image With Circle",image_with_circle)
        #never close the image until any exit button is pressed
        cv2.waitKey(0)
        #after click any exit button it destroy all windows 
        cv2.destroyAllWindows()
        choice_for_save_image=input("Want to save the image (y/n) : ")
        if choice_for_save_image=='y':
            cv2.imwrite("Image_With_Circle.jpg",image_with_circle)
        else:
            print("Ok Thanks.....")
    elif choice=='4':
        x1=int(input("Enter the x1 co-ordinate for the pentagon : "))
        y1=int(input("Enter the y1 co-ordinate for the pentagon : "))
        x2=int(input("Enter the x2 co-ordinate for the pentagon : "))
        y2=int(input("Enter the y2 co-ordinate for the pentagon : "))
        x3=int(input("Enter the x3 co-ordinate for the pentagon : "))
        y3=int(input("Enter the y3 co-ordinate for the pentagon : "))
        x4=int(input("Enter the x4 co-ordinate for the pentagon : "))
        y4=int(input("Enter the y4 co-ordinate for the pentagon : "))
        x5=int(input("Enter the x5 co-ordinate for the pentagon : "))
        y5=int(input("Enter the y5 co-ordinate for the pentagon : "))
        #store the pentagon co-ordinate in array using numpy
        co_of_pentagon = np.array(([x1,y1],[x2,y2],[x3,y3],[x4,y4],[x5,y5]), np.int32)
        #reshape the array
        co_of_pentagon = co_of_pentagon.reshape((-1,1,2))
        #draw a pentagon in image
        image_with_pentagon = cv2.polylines(normal_image, [co_of_pentagon], True, (0,255,255),3)
        cv2.imshow("Image With pentagon",image_with_pentagon)
        #never close the image until any exit button is pressed
        cv2.waitKey(0)
        #after click any exit button it destroy all windows 
        cv2.destroyAllWindows()
        choice_for_save_image=input("Want to save the image (y/n) : ")
        if choice_for_save_image=='y':
            cv2.imwrite("Image_With_Pentagon.jpg",image_with_pentagon)
        else:
            print("Ok Thanks.....")
    elif choice=='5':
        #decide font type
        font_type=cv2.FONT_HERSHEY_SIMPLEX
        x=int(input("Enter the x co-ordinate for the starting point of text : "))
        y=int(input("Enter the y co-ordinate for the starting point of text : "))
        text=input("Enter the text to write on image : ")
        blue_color=int(input("Enter the blue color value for line : "))
        green_color=int(input("Enter the green color  value for line : "))
        red_color=int(input("Enter the red color value for line : "))
        font_size=int(input("Enter the font size : "))
        #write a text in image
        image_with_text=cv2.putText(normal_image,text,(x,y),font_type,font_size,(blue_color,green_color,red_color),cv2.LINE_AA)
        #show the image
        cv2.imshow("Image With Text",image_with_text)
        #never close the image until any exit button is pressed
        cv2.waitKey(0)
        #after click any exit button it destroy all windows 
        cv2.destroyAllWindows()
        choice_for_save_image=input("Want to save the image (y/n) : ")
        if choice_for_save_image=='y':
            cv2.imwrite("Image_With_Text.jpg",image_with_text)
        else:
            print("Ok Thanks.....")
    elif choice=='0':
        print("Thanks for using")
    
```
